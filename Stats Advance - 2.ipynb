```python
#Q1.

# PMF:
#1. Used for discrete random variables.
#2. Gives probability for each possible outcome.

#E.g. Rolling a fair six-sided die: PMF assigns 1/6 to each outcome (1,2,3,4,5,6).

#PDF:
#1. Used for continuous random variables.
#2. Describes likelihood of outcomes in a range.

#E.g. Height of individuals: PDF shows likelihood of different heights without specific probabilities for each height.
```

```python
#Q2.

# CDF:
#1. It gives probability that a random variable is less than or equal to a specified value.
#2. Provides a cumulative view of the probability distribution.

# Why it is used:
#1. Used for finding prababilities in ranges.
#2. Helps calculate median, quartiles, and percentiles.

#Example:
# Suppose we have a continuous random variable representing the time it takes for a customer to be served at a coffee shop.
# If we want to know the probability that a customer is served in under 5 minutes, we can use the CDF to find this cumulative probability.
```

```python
#Q3.

# Examples where the normal distribution is used as a model:
#1. Height of individuals: Heights of people in a population.
#2. Test Scores: Scores on standardized tests like GRE, JEE, GATE, etc.
#3. IQ Scores: Intelligent quotient scores.
#4. Meassurement Errors: Errors in scientific measurements.
#5. Financial Data: Stock returns, asset prices.

# Parameters of the normal distribution:
#1. Mean: Center of the distribution.
#2. Standard Deviation: Spread or width of the distribution.
#3. Variance: Average squared distance from the mean.
#4. Skewness: Measure of asymmetry. A perfectly symmetric distribution has skewness of zero. Positive skewness indicates a tail on the right side, while negative skewness indicates a tail on the left side.
```

```python
#Q4.

# Importance of Normal Distribution:

#1. Simplifies complex data.
#2. Common assumption in statistics.
#3. Facilitates calculations and predictions.
#4. Widely used for modeling real-world phenomena.

# Real-Life Examples of Normal Distribution:

#1. Height of people
#2. Test Scores
#3. Body Weight
#4. Measurement Errors
#5. Income Distribution
#6. Stock Returns
```

```python
#Q5.

# Bernoulli Distribution:

#1. Discrete probability distribution.
#2. Models two outcomes (success/failure).
#3. Parameter: Probability of success (p)

#Example : Coin flip (Heads/Tails)

# Difference b/w Bernoulli and Binomial Distribution:
# Bernoulli: One trial. Binary outcome.
# Binomial: Multiple trials (n). Counts successes in those trials.
# Binomial aggregrates Bernoulli trials.

# Example:
# Bernoulli: Single coin flip (Heads or Tails).
# Binomial: Multiple coin flips, count Heads.
```

```python
#Q7.

# Uniform Distribution:
# A probability distribution where all values withn a specified range are equally likely.
# Probability ;density is constant within the range.

# Example
# A simple example of drawing a random number from 1 to 6 (inclusive) using a fair six-sided die:

# Uniform Distribution: The distribution is uniform because each face of the die has an equal probability of 1/6 of being rolled.

# Characteristics:
# Each number on the die has an equal chance of 1/6 of being rolled.
# The probability density is uniform, with a constant value of 1/6 for each outcome.

#Visual Representation:
# It can be represented using a bar graph 
```

```python
#Q8.

# Z-score is a mesaure of how many standard deviations an observation is away from the mean in a normal distribution.

# Importance of Z-score:

#1. Allows comparison of data from different normal distributions.
#2. Helps identify outliers.
#3. Crucial in hypothesis testing and statistical analysis.
#4. Facilitates making decisions based on relative positions of data points.
```

```python
#Q9.

# Central Limit Theorem:

# A principle stating that, regardless of the population distribution, the sampling distribution of the sample mean approaches a normal distribution as the sample size increases.

# Significance:

# Allows the use of normal distribution assumptions in statistical analysis, even for non-normally distributed populations.
# Foundation for hypothesis testing, confidence intervals, and estimation.
# Widely applied in various fields for data analysis.
```

```python
#Q10.

# Assumptions of the Central Limit Theorem:

#1. Independence: Sampled observations are independent.
#2. Random Sampling: Observations are selected randomly.
#3. Sample size: It should be greater than or equal to 30.
#4. Population Distribution: The population can have any distribution.
```
